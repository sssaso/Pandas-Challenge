```python
import pandas as pd
```

```python
file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#1.Total number of players
total_players = len(purchase_data["SN"].unique())
#Q  totalPlayers_df = pd.DataFrame({"Total Players":total_players})  
# not working --> ValueError: If using all scalar values, you must pass an index
#OK  totalPlayers_df = pd.DataFrame([{"Total Players":total_players}])   <-- list of dict
#OK  totalPlayers_df = pd.DataFrame({"Total Players":total_players}, index=[0])  <-- add index =[0]
totalPlayers_df = pd.DataFrame({"Total Players":[total_players]})  # make variable to LIST !?
totalPlayers_df.head()
```

```python
#2.Number of Unique Items/Average Purchase Price/Total Number of Purchases/Total Revenue
#ok  NumOfUniqItems = len(purchase_data["Item Name"].unique())
UniqItems = purchase_data["Item Name"].unique()
AvgPurPrice = purchase_data["Price"].mean()
NumOfPurchases = len(purchase_data["Purchase ID"])
totalRev = purchase_data["Price"].sum()

Purchasing_Analysis_df =pd.DataFrame({"Number of Unique Items":len(UniqItems),
                                    "Average Price":"${:,.2f}".format(AvgPurPrice),
                                    "Number of Purchases":NumOfPurchases,
                                    "Total Revenue":"${:,.2f}".format(totalRev)}, index =[0])

#ok Purchasing_Analysis_df["Average Price"] = Purchasing_Analysis_df["Average Price"].map("${:,.2f}".format)
#ok Purchasing_Analysis_df["Total Revenue"] = Purchasing_Analysis_df["Total Revenue"].map("${:,.2f}".format)
Purchasing_Analysis_df
```

```python
#Male players / Female players / Others
male_df=purchase_data.loc[purchase_data["Gender"]=="Male"]
female_df=purchase_data.loc[purchase_data["Gender"]=="Female"]
others_df =purchase_data.loc[purchase_data["Gender"]=="Other / Non-Disclosed"]

malePlayers = len(male_df["SN"].unique())
femalePlayers = len(female_df["SN"].unique())
others = len(others_df["SN"].unique())

malePerc = "{:.2%}".format(malePlayers/total_players)
femalePerc = "{:.2%}".format(femalePlayers/total_players)
othersPerc = "{:.2%}".format(others/total_players)

genderDemog_df = pd.DataFrame({"Total Count":[malePlayers,femalePlayers, others], 
                              "Percentage of Players":[malePerc, femalePerc, othersPerc]}, 
                             index=["Male","Female", "Other/Non-Disclosed"])
genderDemog_df
```

```python
#  * Purchase Count
purchase_data['Count'] = 1
Purchase_Count = purchase_data.groupby(["Gender"]).count()['Count']
purchaseCount_male = Purchase_Count["Male"]
purchaseCount_female = Purchase_Count["Female"]
purchaseCount_others =Purchase_Count["Other / Non-Disclosed"]

#  * Average Purchase Price
avg_Purchase_price = purchase_data.groupby(["Gender"]).mean()['Price']
avg_Purchase_price_male = avg_Purchase_price["Male"]
avg_Purchase_price_female = avg_Purchase_price["Female"]
avg_Purchase_price_others = avg_Purchase_price["Other / Non-Disclosed"]

#  * Total Purchase Value
tot_Purchase_value = purchase_data.groupby(["Gender"]).sum()['Price']
tot_Purchase_value_male = tot_Purchase_value["Male"]                      
tot_Purchase_value_female = tot_Purchase_value["Female"]  
tot_Purchase_value_others = tot_Purchase_value["Other / Non-Disclosed"]

#  * Average Purchase Total per Person by Gender
avg_tot_pur_male = tot_Purchase_value_male / malePlayers
avg_tot_pur_female = tot_Purchase_value_female / femalePlayers
avg_tot_pur_others = tot_Purchase_value_others / others

gender_purchase_df = pd.DataFrame({"Purchase Count":[purchaseCount_male,purchaseCount_female, purchaseCount_others], 
                              "Average Purchase Price":["${:.2f}".format(avg_Purchase_price_male), "${:.2f}".format(avg_Purchase_price_female), "${:.2f}".format(avg_Purchase_price_others)],
                               "Total Purchase Value":["${:,.2f}".format(tot_Purchase_value_male),"${:,.2f}".format(tot_Purchase_value_female),"${:,.2f}".format(tot_Purchase_value_others)],
                                 "Avg Total Purchase per Person":["${:.2f}".format(avg_tot_pur_male),"${:.2f}".format(avg_tot_pur_female),"${:.2f}".format(avg_tot_pur_others)]
                                  }, 
                             index=["Male","Female", "Other/Non-Disclosed"])
gender_purchase_df
```
